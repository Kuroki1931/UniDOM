```python
import io
import cv2
import numpy as np
import pickle
import matplotlib.pyplot as plt
import random
from mpl_toolkits.mplot3d import axes3d, Axes3D

import glob
```

```python
with open('/root/DaXBench/logs/apg_para/fold_cloth1_para/fold_cloth1_para_ep_len3_num_envs2_lr0.0001_max_it2000_max_grad_norm0.3/seed0/apg_reward_dict.pkl', 'rb') as f:
    data_para = pickle.load(f)
with open('/root/DaXBench/logs/apg/fold_cloth1/fold_cloth1_ep_len3_num_envs2_lr0.0001_max_it2000_max_grad_norm0.3/seed0/apg_reward_dict.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
def dict_to_list(dict):
    key_list = []
    value_list = []
    stiff_list = []
    for key in dict:
        key_list.append(key)
        value_list.append(dict[key][0])
        stiff_list.append(dict[key][1])
        
    return key_list, value_list, stiff_list
```

```python
key_list, value_list, stiff_list = dict_to_list(data)
key_para_list, value_para_list, stiff_para_list = dict_to_list(data_para)
```

```python
fig, ax1 = plt.subplots()

# Plot the line plot on the left y-axis
ax1.plot(key_list, value_list, 'g')
ax1.plot(key_list, value_para_list, 'r')
ax1.set_xlabel('X data')
ax1.set_ylabel('reward', color='g')
ax1.tick_params('y', colors='g')

# Create a new set of axes for the right y-axis
ax2 = ax1.twinx()

# Plot the bar plot on the right y-axis
ax2.bar(key_list, stiff_para_list, width=20, alpha=0.7)
ax2.set_ylabel('stiff', color='b')
ax2.tick_params('y', colors='b')
ax2.set_ylim([0, max(key_list)])

plt.show()
```
