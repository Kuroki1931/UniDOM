```python
def dict_to_list(dict):
    key_list = []
    value_list = []
    for key in dict:
        key_list.append(key)
        value_list.append(dict[key][0])
        
    return key_list, value_list 
```

```python
key_list, value_list = dict_to_list(data)
key_para_list, value_para_list = dict_to_list(data_para)
```

```python
plt.plot(key_list, value_list, label='apg', color='red')
plt.plot(key_list, value_para_list, label='apg', color='blue')
```
